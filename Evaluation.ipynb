## Extract references

```python
def extract_ref (function, fq_type, all_output):
    """
    extract corresponding references based on function and fq_type
    
    Return:::
    f_ref_lss, f_cad_ls for the evaluation
    
    Para:::
    all_output is overall output of function and fq_types
    """
    refe_path = f"response_data/references/filtered/{function}.json"
    with open(refe_path, "r") as refer:
        refer_dics = json.load(refer)
        
 # prepare the evaluation form, extract question utterance
    if fq_type == "function":
        ref_all = [dic.get('FQ on function level') for dic in refer_dics]
        ref_lss= [[[s] for s in clean_str(refs)] for refs in ref_all]
        
    else:
        ref_all = [dic.get('FQ on feedback') for dic in refer_dics]
        ref_lss = [[[s] for s in clean_str(refs)] for refs in ref_all]
    
    print(f"original reference of {function}_{fq_type} amount:{len(ref_lss)}")
    
    
    #extract candidate list
    cad_ls = all_output[function][fq_type]['output']
#     check amount
    if not len(cad_ls) == len(ref_lss):
        print("candidates amount not match references")
        
#     filter potential empty references   
    f_ref_lss, f_cad_ls, conver_ind= [],[],[]

    for n, (ref_ls, cad_str) in enumerate(zip (ref_lss, cad_ls)):
        if ref_ls:
            f_ref_lss.append(ref_ls)
            f_cad_ls.append(cad_str)
        else:
            print()
            conver_ind.append(n)

        # print(ref_ls,emo_str)
        # break
    if not len(f_ref_lss) == len(f_cad_ls):
        print("filtered candidates amount not match references")
        
    print(f"Filtered reference of {function}_{fq_type} amount: {len(f_ref_lss)};\n"
      f"Removed conversation index: {conver_ind}")
    
    return(f_ref_lss, f_cad_ls)
    
```

## Fine-tuned Zero-shot scores

```python
ftoutput_0 = "response_data/test_data/out_fq/out_clean/ftoutput_zero_clean.json"
with open(ftoutput_0, "r") as f0:
    ftd_0 = json.load(f0)
    
# print(len(d_0["communication"]["function"]["output"]))

# function = "communication"
# fq_type = "function"#or emotion

# ref_lss, cad_ls = extract_ref (function, fq_type, d_0)

# print(len(cad_ls))
# print(cad_ls[:2])
```

```python
functions = ["self-care", "communication", "mobility"]
fq_types = ["emotion", "function"]

for function in functions:
    for fq_type in fq_types:
        ref_lss, cad_ls = extract_ref(function, fq_type, ftd_0)
        
        # Scores of each conversation
        scores_ave = evaluate_conversations_new(cad_ls, ref_lss, cal_high=False)
        ftd_0[function][fq_type]["score_per"] = scores_ave
        
        # Average score of all conversations for fq_type/function
        ave_type_0 = ave_type_cate(scores_ave)
        ftd_0[function][fq_type]["average_score"] = ave_type_0  # Store the average separately

        
```
